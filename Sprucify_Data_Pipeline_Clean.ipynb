#### Aggregate Reviews and Ratings 

```python
## url constructor
pagination = ['','?start=40','?start=80']
business = ['biz/j-p-kempt-barber-san-francisco']
url_base  = 'http://www.yelp.com/'

## generate url
url = [ url_base + base + page for base in business for page in pagination ] 
print url
```

```python
def process_reviews(soup):
    review_body = str(soup.find_all(class_ ='review_comment ieSucks'))
    processed_review_body = re.split('</?\w*\s[^>]*>', review_body)[1:]
    processed_review_body = [re.sub('</?[a-z]{1,2}>','', i) for i in processed_review_body] 
    
    return processed_review_body

def process_rating(soup):
    ratings = str(soup.find_all(class_ ='rating-very-large'))
    process_ratings = re.split('<div class="rating-very-large">', ratings)[1:]
    rate = []
    for i in range(len(process_ratings)):
        anchor = process_ratings[i].find('title="')
        rate.append(process_ratings[i][anchor+7:anchor+10])
    
    # index - 0 review is the barbers review
    return rate[1:]



provider_reviews = []
provider_ratings = []

for page in url:
    r = requests.get(page)
    if r.status_code == 200:
        print '===Status OK===='
        
        # get the html content of the page
        html_body = r.content
        soup = BeautifulSoup(html_body, 'html.parser')
        provider_reviews.extend(process_reviews(soup))
        provider_ratings.extend(process_rating(soup))

        
        # take a rest 
        time.sleep(1)
        
        
print len(provider_reviews)
print len(provider_ratings)
```
